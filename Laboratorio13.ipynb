Prompt 1: Análisis de Sentimiento con TextBlob

Instrucciones:
Utiliza la librería TextBlob para analizar el sentimiento de las siguientes frases. Debes interpretar el sentimiento de las frases como positivo, negativo o neutral en base a su polaridad.

Frase 1: "The new design significantly improved user satisfaction and boosted our sales."
Frase 2: "The system performance was disappointing and caused multiple delays in the project."
Frase 3: "The technical tests were completed as planned and met the basic requirements."

Objetivo:
Para cada frase, analiza la polaridad y asigna el sentimiento correspondiente:

Positivo si la polaridad es mayor que 0.
Negativo si la polaridad es menor que 0.
Neutral si la polaridad es 0.

```python
from textblob import TextBlob

# Lista de frases
frases = [
    "The new design significantly improved user satisfaction and boosted our sales.",
    "The system performance was disappointing and caused multiple delays in the project.",
    "The technical tests were completed as planned and met the basic requirements."
]

# Función para clasificar el sentimiento
def clasificar_sentimiento(polaridad):
    if polaridad > 0:
        return "Positivo"
    elif polaridad < 0:
        return "Negativo"
    else:
        return "Neutral"

# Análisis
for i, frase in enumerate(frases, 1):
    blob = TextBlob(frase)
    polaridad = blob.sentiment.polarity
    sentimiento = clasificar_sentimiento(polaridad)

    print(f"Frase {i}: {frase}")
    print(f"Polaridad: {polaridad}")
    print(f"Sentimiento: {sentimiento}")
    print("-" * 50)
```

Frase 1: "The new design significantly improved user satisfaction and boosted our sales."
Traducción: "El nuevo diseño mejoró significativamente la satisfacción del usuario y aumentó nuestras ventas."
Frase 2: "The system performance was disappointing and caused multiple delays in the project."
Traducción: "El rendimiento del sistema fue decepcionante y causó múltiples retrasos en el proyecto."
Frase 3: "The technical tests were completed as planned and met the basic requirements."
Traducción: "Las pruebas técnicas se completaron según lo planeado y cumplieron los requisitos básicos."

Prompt 2: Entrenamiento de Modelo de Clasificación de Sentimientos con Naive Bayes

Instrucciones:
Entrena un modelo de clasificación de sentimientos utilizando el algoritmo Naive Bayes. Utiliza los siguientes ejemplos de frases etiquetadas y evalúa la precisión de tu modelo.

Datos de ejemplo:

Frase 1: "This app is very intuitive!"
Frase 2: "I didn't enjoy the customer service."
Frase 3: "Quite average experience."
Frase 4: "Fantastic presentation!"
Frase 5: "The food was awful."

Objetivo:

Entrena un modelo de Naive Bayes con estas frases etiquetadas.
Evalúa la precisión del modelo utilizando las frases proporcionadas.

```python
import textblob
from textblob.classifiers import NaiveBayesClassifier
import nltk

# Descargar los datos necesarios para NLTK (utilizado por TextBlob)
try:
    nltk.download('punkt_tab')
    nltk.download('averaged_perceptron_tagger') # Often needed by TextBlob
except Exception as e:
    print(f"Could not download NLTK corpora: {e}. Assuming they are already present or network issue.")

# Datos etiquetados (texto, etiqueta)
train_data = [
    ("This app is very intuitive!", "pos"),
    ("I didn't enjoy the customer service.", "neg"),
    ("Quite average experience.", "neu"),
    ("Fantastic presentation!", "pos"),
    ("The food was awful.", "neg")
]

# Crear y entrenar el clasificador
cl = NaiveBayesClassifier(train_data)

# Evaluación de precisión con los mismos datos
accuracy = cl.accuracy(train_data)

print(f"Precisión del modelo: {accuracy:.2f}")
print("-" * 50)

# Probar predicciones
for frase, etiqueta_real in train_data:
    prediccion = cl.classify(frase)
    print(f"Frase: {frase}")
    print(f"Etiqueta real: {etiqueta_real}")
    print(f"Predicción: {prediccion}")
    print("-" * 50)
```
